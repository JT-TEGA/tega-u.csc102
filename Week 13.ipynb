```python
import pandas as pd
from abc import ABC, abstractmethod

class PAU_SIMS(ABC):
    @abstractmethod
    def info(self):
        pass
    
class Student1(PAU_SIMS):
    def info(self,data):
        self.data = ({'Name':["Oluchi Mordi"],'Matric Number':["ACC10211111"],'Department':["Accounting"],'Level':[300]})
        df =pd.DataFrame(self.data)
        print("This is student info")
        df.to_csv('Student1.csv')
        print(df)

class Student2(PAU_SIMS):
    def info(self,data):
        self.data = ({'Name':["Adams Aliyu"],'Matric Number':["ECO10110101"],'Department':["Economics"],'Level':[100]})
        df =pd.DataFrame(self.data)
        print("This is student info")
        df.to_csv('Student2.csv')
        print(df)
        

class Student3(PAU_SIMS):
    def info(self,data):
        self.data = ({'Name':["Shanai Bolade "],'Matric Number':["CSC10328828"],'Department':["Computer"],'Level':[200]})
        df =pd.DataFrame(self.data)
        print("This is student info")
        df.to_csv('Student3.csv')
        print(df)
        
class Student4(PAU_SIMS):
    def info(self,data):
        self.data = ({'Name':["Adekunle Gold"],'Matric Number':["EEE11020202"],'Department':["Electrical"],'Level':[200]})
        df =pd.DataFrame(self.data)
        print("This is student info")
        df.to_csv('Student4.csv')
        print(df)
        
class Student5(PAU_SIMS):
    def info(self,data):
        self.data = ({'Name':["Blanca Edemoh"],'Matric Number':["MEE10202001"],'Department':["Mechanical"],'Level':[100]})
        df =pd.DataFrame(self.data)
        print("This is student info")
        df.to_csv('Student5.csv')
        print(df)
        
        
choice = input("Which Student File would you like to access?:")
if choice == "student1".lower():
    csc = Student1()
    csc.info(choice)
elif choice == "student2".lower():
    csc = Student2()
    csc.info(choice)
elif choice == "student3".lower():
    csc = Student3()
    csc.info(choice)
elif choice == "student4".lower():
    csc = Student4()
    csc.info(choice)
elif choice == "student5".lower():
    csc = Student5()
    csc.info(choice)
else:
    print("Student not recognized")
```

```python
import pandas as pd

assests = [1000000, 1500000, 3000000, 250000, 800000]
liabilities = [200000, 500000, 1500000, 50000, 300000]
percentage_leverage_list = []

for i in range(5):
    percentage_leverage = (assests[i] - liabilities[i]) / (assests[i]) * 100
    percentage_leverage = round(percentage_leverage)
    percentage_leverage_list.append(percentage_leverage)
    
df =   pd.DataFrame({"Company" : ["Enron", "Anderson", "GK Jones", "Mica", "Dune"],
                     "Founded": ["1987", "1936", "2001", "1996", "2008"],
                     "Assests": [1000000, 1500000, 3000000, 250000, 800000],
                     "Liabilities": [200000, 500000, 1500000, 50000, 300000],
                     "Percentage Leverages (%)": [x for x in percentage_leverage_list]
                    })
df.to_csv("Company data.csv")
df
```
